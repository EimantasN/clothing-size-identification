```python
df = pd.read_csv("./dataset/dataset.csv")
```

```python
path = "./dataset/imgs/"
imgs = [f for f in listdir(path) if isfile(join(path, f))]

for i in range(len(imgs)):
    if ".JPG" in  imgs[i]:
        img = cv2.imread("./dataset/imgs/" + imgs[i])
        cv2.imwrite("./dataset/imgs/" + imgs[i].split(".")[0] + ".jpg", img)
```
